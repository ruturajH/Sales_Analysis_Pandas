```python
import pandas as pd
import os
```

#### Task 1: Merge 12 Months of Sales Data into a single file

```python
#Start with one csv file
df = pd.read_csv("./Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data/Sales_May_2019.csv")

files = [file for file in os.listdir('./Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data')]

#Create a new Data Frame
all_month_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data/"+file)
    all_month_data = pd.concat([all_month_data, df])

all_month_data.to_csv("all_data.csv", index = False)
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

## Clean Up the Data!

```python
all_data.head(25)
```

Drop NAN rows

```python
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df
```

```python
all_data = all_data.dropna(how = 'all')
all_data.head()
```

#### FInd 'OR' and deleting it

```python
temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
temp_df
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

#### Convert Columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])   #Make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) #Make float

all_data.head()
```

#### Task 2: Add Month Column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head(25)
```

#### Task 3 : Add a Sales Columns

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Task 4: Add a city column

```python
# Apply method

all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

all_data.head()
```

#### Question 1: What was the best month for sales? How much was earned during that month?

```python
result = all_data.groupby('Month').sum()
result
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,result['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Month')
plt.show()
```

#### Best month for Sales was December with approx 4mn sales and worst month was January with 2.8mn of sales.

#### Question 2: Which city had the highest number of sales?

```python
result = all_data.groupby('City').sum()
result
```

```python
import matplotlib.pyplot as plt

City = [city for city, df in all_data.groupby(['City'])]

plt.bar(City,result['Sales'])
plt.xticks(City, rotation = 'vertical',size = 9)
plt.ylabel('Sales in USD')
plt.xlabel('City')
plt.show()
```

#### Question 3: What time should we display advertisements to maximize likelihood of customer's buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Count of order')
plt.grid()
plt.show()
```

#### The best time to adverstise is around 11am (11) or 6pm (18)

#### Question 4: What products are most often sold together?

```python
all_data.groupby('Product').sum()
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID','Grouped']].drop_duplicates()

df.head(50)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df["Grouped"]:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key,value in count.most_common(10):
    print(key,value)
```
